```python
species=iris['Species']
species
```

```python
species_col =  iris['Species'].map({'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2  })
species_col
```

```python
iris_data = [(p_length, p_width) for p_length, p_width in zip(p_length, p_width) ]
iris_data
```

```python
iris = [(p_length, p_width,species_col) for p_length, p_width,species_col in zip(p_length, p_width,species_col) ]
iris
```
