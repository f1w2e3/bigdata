```python
import numpy as np
import pandas as pd

#특성 선택하기. Age, Fare
Age=train_dt['Age']
Fare=train_dt['Fare']


# Age 컬럼에서 결측치가 있는 인덱스를 A 배열에 저장
A = train_dt[train_dt['Age'].isnull()].index.to_numpy()

# Fare 컬럼에서 결측치가 있는 인덱스를 B 배열에 저장
B = train_dt[train_dt['Fare'].isnull()].index.to_numpy()

# A와 B 배열을 합치고 중복된 인덱스는 1번만 나타나도록 처리
C = np.unique(np.concatenate((A, B)))

print(f"Age 결측치 인덱스: {A}")
print(f"Fare 결측치 인덱스: {B}")
print(f"합쳐진 인덱스: {C}")

Age_filtered = Age[~np.isin(np.arange(len(Age)), C)]
Fare_filtered = Fare[~np.isin(np.arange(len(Fare)), C)]
print(Age_filtered)
print(Fare_filtered)


#People=Age+Fare 이렇게 하면 안되는 이유: 2차원 배열이 아니라 1차원으로 쭉 연결됨
People_data=np.column_stack((Age_filtered,Fare_filtered))

print(People_data)

People_target = train_dt['Survived'].to_numpy()
#생존자 담은 명단 배열
People_target_filtered = People_target[~np.isin(np.arange(len(People_target)), C)]
#결측치 처리
print(People_target_filtered)

import matplotlib.pyplot as plt
plt.scatter(Age_filtered,Fare_filtered)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    People_data, People_target_filtered, random_state=42)

print(train_input.shape, train_target.shape)
#shape로 데이터 형태 확인하기
print(test_input.shape, test_target.shape)


from sklearn.neighbors import KNeighborsClassifier

kn=KNeighborsClassifier()

kn.fit(train_input, train_target)
#학습시키기

kn.score(test_input, test_target)
#점수 표시

kn.predict(test_input)
#테스트 인풋을 예측한 값 표시
```

Fare_filtered가 pandas Series이고 People_target_filtered가 NumPy 배열이라면, Fare_filtered[737]과 People_target_filtered[737]은 의미가 다릅니다. KeyError가 발생하는 이유와 해결 방법을 설명드리겠습니다.

1. Fare_filtered[737] (pandas Series):

pandas Series에서 [] 연산자는 label-based indexing (라벨 기반 인덱싱)과 integer position-based indexing (정수 위치 기반 인덱싱)을 모두 지원합니다.

Label-based indexing: Series의 인덱스 라벨을 사용하여 값에 접근합니다. Fare_filtered의 인덱스가 숫자로 되어 있더라도, pandas는 이를 라벨로 취급합니다. Fare_filtered[737]은 인덱스 라벨이 737인 값을 찾으려고 시도합니다. 만약 인덱스 라벨 737이 없다면 KeyError가 발생합니다.
Integer position-based indexing (.iloc 사용): 정수 위치를 사용하여 값에 접근하려면 .iloc 접근자를 사용해야 합니다. Fare_filtered.iloc[737]은 737번째 위치에 있는 값을 반환합니다. 이 경우 인덱스 라벨과는 관계없이 실제 데이터의 위치를 기반으로 값에 접근합니다. 만약 737번째 위치가 존재하지 않으면 IndexError가 발생합니다.
